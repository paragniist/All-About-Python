<h3>What is a Program?</h3>
A program is a sequence of instructions written in a programming language. Programming languages can be anything; in this course, we focus on Python. Examples of programming languages include Python, C, and C++.

A program consists of instructions that get executed based on how the program is written.

For example, Google Chrome is an executable file (EXE) and an application. It is a sequence of instructions written in some programming language that defines how the browser should work. Similarly, applications like Microsoft Word and Excel are also programs.

<h3>What is a Process?</h3>
A process is an instance of a program that is being executed. For example, when you double-click Google Chrome, a process starts and the browser opens. You can open multiple instances of the browser, and each instance is a separate process.

A process works along with the operating system and requires various computer resources. These resources include:

Code Segment: Contains the program code required to run the process.
Data Segment: Contains global and static variables accessible throughout the program.
Heap Memory: Used for dynamic memory allocation during execution.
Stack: Stores local variables and function call information.
Registers: Small memory locations used to store temporary variables and logic for short periods.
Each process has its own separate memory space, which prevents one process from corrupting another. For example, if one Google Chrome browser instance crashes, it does not affect other instances.

However, because each process has separate memory, switching between processes can increase execution time.

Examples of Processes
Common examples of processes include:

Multiple instances of a browser like Google Chrome.
Opening Excel sheets.
Running various applications installed on your system.
Each process has a unique process ID (PID) assigned when it is launched.

You can observe running processes in your system's task manager, which shows applications like Calculator, Google Chrome, Notepad++, OBS Studio, and background processes along with their CPU, memory, and disk usage.

<h3>What is a Thread?</h3>
A thread is a unit of execution within a process. Threads share the code segment, data segment, and heap memory of the process but have their own stack and registers.

<h4>Single-Threaded Process</h4>
In a single-threaded process, there is only one thread executing. This thread has its own stack and registers but shares the rest of the process's memory.

Example of Threads
Consider the application MSPaint:

The MSPaint application runs as a process.
When you create a rectangular box, a thread is created within the process to perform this task.
Similarly, creating a circular box involves creating another thread.
Each thread handles specific tasks within the process.

Similarly, in Excel, selecting an option like making text bold creates a thread within the process to perform that operation.

<h4>Multi-Threaded Process</h4>
In a multi-threaded process, multiple threads execute concurrently within the same process. Each thread has its own stack and registers but shares the code, data, and heap segments with other threads.

For example, if a process has two threads, each thread executes independently with its own stack and registers while sharing the rest of the process's resources.

This allows for concurrent execution of tasks within a single process, improving efficiency and responsiveness.

Summary
A program is a sequence of instructions written in a programming language.
A process is an executing instance of a program with its own separate memory and resources.
A thread is a unit of execution within a process, sharing some resources but having its own stack and registers.
Processes are isolated from each other, preventing corruption.
Threads within a process enable concurrent execution of tasks.
Understanding these concepts is essential for working with multithreading and multiprocessing in Python and other programming languages.

<h4>Key Takeaways</h4>
A program is a sequence of instructions written in a programming language, such as Python or C.
A process is an instance of a program that is being executed, with its own separate memory space and resources.
Processes have distinct code, data, heap, stack, and register segments, ensuring isolation and preventing corruption between processes.
A thread is a unit of execution within a process, sharing code, data, and heap segments but having its own stack and registers.
Single-threaded processes have one thread, while multi-threaded processes have multiple threads executing concurrently within the same process.
Understanding processes and threads is fundamental for writing efficient applications using multithreading and multiprocessing.
