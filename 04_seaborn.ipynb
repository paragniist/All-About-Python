```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
## basic plotting with seaborn

tips = sns.load_dataset('tips')
tips
```

```python
## creating a scatterplot

sns.scatterplot(x='total_bill',y='tip',data=tips,color='red')
plt.title("Scatter plot of Total Bill")
plt.show
```

```python
## LINE PLOT FOR THE SAME DATA

sns.lineplot(x='size',y='total_bill',data=tips)
```

```python
## CATEGORICAL PLOTS

## BOX PLOT

sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
## VIOLIN PLOT

sns.violinplot(x='day',y='total_bill',data=tips, color='pink')
```

```python
## HISTOGRAM

sns.histplot(tips['total_bill'],bins=10,kde=True)
```

```python
## KDE PLOT

sns.kdeplot(tips['total_bill'],fill=True)
```

<h3> PAIR PLOT -</h3><h4>Shows relation between each and every variable:

```python
## PAIR PLOT 

sns.pairplot(tips)
sns.catplot(tips)
```

```python
## HEAT MAP:
corr = tips[['total_bill','tip','size']].corr()
corr

## because data table conists of Gender variable which is string type
```

```python
sns.heatmap(corr, cmap="coolwarm", annot=True)
```

<h3>Q. USING MATPLOTLIB TO READ DATA AND USING SEABORN TO PLOT BAR BLOT:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Sales Data.csv')
df.head(5)
```

```python
plt.figure(figsize=(20,10))
sns.barplot(data=df,x='Product',y='Sales',estimator=np.mean,edgecolor='black')
plt.title('Price of Each Product')
plt.xlabel('Products')
plt.ylabel('Sales')
```
